```python
# Pick up address and ward and lat,long from geoaddress from City of toronto 
df2 = pd.read_csv('GeoAddress.csv')
```

```python
df2.head()
```

```python
df2.columns
```

```python
# Drop the columns we don't need. 
# We just want lat and long info ward and address. 
# Later on we can enhance with postcode 
#
df2.drop(columns = [ 'GEO_ID', 'LINK', 'MAINT_STAG',  'LONUM',
       'LONUMSUF', 'HINUM', 'HINUMSUF', 'ARC_SIDE', 'DISTANCE', 'FCODE',
       'FCODE_DES', 'CLASS', 'NAME', 'X', 'Y', 
       'OBJECTID',  'geometry'  ], axis=1, inplace=True)

# check the columns 
df2.head()
```

```python
# Remove any spaces leading or trailing  before merging with the parking ticket dataframe

df2['LFNAME'] = df2['LFNAME'].str.replace(r'\s+', ' ')
df2['ADDRESS'] = df2['ADDRESS'].str.replace(r'\s+', ' ')
```

```python
# Convert to strings so that I can concatenate them in same format as the parking ticket dataframe.
# I want to do a lookup for the wards on the Queen st dataframe and then merge the two dataframes

# Format df2['Location'] = df2['streetnum']  + df2['street']

df2['ADDRESS'] = df2['ADDRESS'].astype(str)
df2['WardLocation'] = df2['ADDRESS']  +' '+  df2['LFNAME']
```

```python
# Upper case the address as per the parking ticket dataframe

df2['WardLocation'] = df2['WardLocation'].str.upper()
```

```python
df2.head()
```

```python
df.head()
```

```python
df2.columns
```

```python
# Create a merged dataframe, with all parking tickets, and a lookup by street name and number 
# to pick up the lat, long and ward info

merged_df = pd.merge(df[['date_of_infraction', 'infraction_code', 'infraction_description','set_fine_amount', 'time_of_infraction', 'location2'   ]],
                     df2[[ 'LONGITUDE', 'LATITUDE','MUN_NAME', 'WARD_NAME', 'WardLocation']],
                     left_on = 'location2', 
                     right_on = 'WardLocation', 
                     how='left')

print (merged_df)
```
