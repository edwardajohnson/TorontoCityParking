### Convert city data from a Excel to processing for geocoding
The geocoding will be shown in another script<br>
The data is provided in a multi-sheeted Excel file<br>

The output will be used to expand out address, location and neighbourood information on every single street proprty in Toronto.


```python
import pandas as pd

# Header rows are pre-removed
df = pd.concat(pd.read_excel('TorontoStreetWards.xlsx', sheet_name=None, skiprows=0,sort=True))
```

```python
# Check the size and format of the DataFrame 
df.columns
```

```python
df.head(5)
```

#### The data is formatted as two main columns each with street, street numbers, ward and subward components 

```python
df.columns = ['StreetA','OddA1','OddA2','EvA1','EvA2','WardA','SubwardA',
            'StreetB','extra1','OddbB','OddB2','EvB1','EvB2','WardB', 'SubwardB', 'extra2' ]

# We don't need the extra columns of blank space 
df.drop(columns = ['extra1','extra2'], inplace=True)
```
